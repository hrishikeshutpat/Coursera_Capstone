```python
def generate_location_data(html_data):
    #---Reading Scrapped HTML Content
    soup = BeautifulSoup(html_content, 'html')
    table_data = soup.find("table")
    table_rows = table_data.find_all("tr")
    table_contents = table_data.find_all("td")
    contents = []
    for tc in table_contents:
        contents.append(tc.string.replace('\n', ''))
    #---Generating Dataframe & Cleaning Data---
    city_location_data = []
    for n in range(1, int(len(contents)/4)):
       city_location_data.append(contents[(4*n):4*(n+1)])
    city_location_data = pd.DataFrame(city_location_data, columns = ['Location', 'Pincode', 'State', 'District'])
    
    return(city_location_data)
```

<h3>V] Transforming Venue Data</h3>

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
def return_city_grouped(city_venue):
    city_onehot = pd.get_dummies(city_venue[['Venue Category']], prefix="", prefix_sep="") #---one hot encoding
    city_onehot['Neighborhood'] = city_venue['Neighborhood'] #---add neighborhood column back to dataframe
    fixed_columns = [city_onehot.columns[-1]] + list(city_onehot.columns[:-1]) #---move neighborhood column to the first column
    city_onehot = city_onehot[fixed_columns] #---Group Neighborhoods & Find Mean Frequency
    city_grouped = city_onehot.groupby('Neighborhood').mean().reset_index()
    
    return(city_grouped)
```

```python
def return_venue_sort_data(city_grouped):
    #---Step 1---
    num_top_venues = 5
    for hood in city_grouped['Neighborhood']:
        #print("----"+hood+"----")
        temp = city_grouped[city_grouped['Neighborhood'] == hood].T.reset_index()
        temp.columns = ['venue','freq']
        temp = temp.iloc[1:]
        temp['freq'] = temp['freq'].astype(float)
        temp = temp.round({'freq': 2})
        #print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
        #print('\n')

    #---Step 2---
    num_top_venues = 10
    indicators = ['st', 'nd', 'rd']
    # create columns according to number of top venues
    columns = ['Neighborhood']
    for ind in np.arange(num_top_venues):
        try:
            columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
        except:
            columns.append('{}th Most Common Venue'.format(ind+1))
            
    # create a new dataframe
    neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
    neighborhoods_venues_sorted['Neighborhood'] = city_grouped['Neighborhood']

    for ind in np.arange(city_grouped.shape[0]):
        neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(city_grouped.iloc[ind, :], num_top_venues)

    return(neighborhoods_venues_sorted)
```

<h3>VI] Clustering</h3>

In this section, the neighborhoods of each city, as identified by their pincodes, shall be clustered using the KMM Algorithm.

```python
def city_clustering(city_grouped, city_venue_sorted, city_venue):  
    kclusters = 5
    city_grouped_clustering = city_grouped.drop('Neighborhood', 1)
    kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(city_grouped_clustering)
    kmeans.labels_[0:10]
    city_venue_sorted.insert(0, 'Cluster Labels', kmeans.labels_)
    city_merged = city_venue
    city_merged = city_merged.join(city_venue_sorted.set_index('Neighborhood'), on='Neighborhood')
    
    return(city_merged)
```

<h3>VII] Plotting</h3>


In this final section, the clusters shall be mapped onto the cities using the Folium Library

```python
def plot_map(city_merged):
    latitude = city_merged.loc[0]['Neighborhood Latitude']
    longitude = city_merged.loc[0]['Neighborhood Longitude']
    kclusters = 5
    map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)
    x = np.arange(kclusters)
    ys = [i + x + (i*x)**2 for i in range(kclusters)]
    colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
    rainbow = [colors.rgb2hex(i) for i in colors_array]
    markers_colors = []
    for lat, lon, poi, cluster in zip(city_merged['Neighborhood Latitude'], city_merged['Neighborhood Longitude'], city_merged['Neighborhood'], city_merged['Cluster Labels']):
        label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
        folium.CircleMarker(
            [lat, lon],
            radius=5,
            popup=label,
            color=rainbow[cluster-1],
            fill=True,
            fill_color=rainbow[cluster-1],
            fill_opacity=0.7).add_to(map_clusters)

    return(map_clusters)
```
